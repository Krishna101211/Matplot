#**Bar Chart**

```python
import matplotlib.pyplot as plt
names=["Python","Java","Html","CSS","SQL"]
rank=[85,70,50,65,45]
# plt.bar(names,rank)
plt.bar(names,rank,width=0.3,color=["r","m","c","y","pink"],label=names,edgecolor="black",linestyle="solid",
        alpha=0.5)
plt.title("Krishi_Bar",fontsize=20,color="red")
plt.xlabel("X-axis",fontsize=10,color="blue")
plt.ylabel("Y-axis",fontsize=10,color="green")
plt.legend(loc="upper right",framealpha=True,shadow=True,facecolor="yellow"
           ,edgecolor="black",title="BAR_GRAPH",title_fontsize=15,fancybox=True)
plt.grid(axis="y")
plt.show()
```

```python
names=["python","java","C","C++","#"]
rank=[10,20,30,40,50]

plt.bar(names,rank,
        width=0.3,#use to reduce width
        color=["r","b","y"],
        edgecolor="black",
        # facecolor="Black",
        )
plt.legend(loc='upper left',edgecolor='black',shadow=True,fancybox=True)
plt.label=names
plt.xlabel("X-axis",fontsize=20,color='blue')
plt.ylabel("Y-axis",fontsize=20)
plt.grid()
plt.show()
```

```python
from matplotlib import legend
import numpy as np

players=["Dhoni","virat","Rohit","HP"]
Ipl_score=[400,200,300,400]
DLF_score=[350,300,100,100]
Tata_ipl=[100,200,200,300]
width=0.3

x=np.arange(len(players))
print(x)
y=[i+width for i in x]
print(y)

z=[i+width for i in y]
print(z)

plt.bar(x,Ipl_score,width=0.3,color="r",edgecolor="black")
plt.bar(y,DLF_score,width=0.3)


plt.legend()
plt.xticks(x+width/2,players)#this make show labels of player is between two bars
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
names=["virat","Rahul","HP","Dhoni","Rohit"]
IPL_Score=[400,350,700,200,600]
DLF_Score=[350,200,250,300,500]
TATA_IPL=[200,150,100,200,250]
width=0.3
x=np.arange(len(names))
print(x)
y=[i+width for i in x]
print(y)

z=[i+width for i in y]
print(z)

plt.bar(x,IPL_Score,width=width,color="red",edgecolor="black",label="IPL_Score")
plt.bar(y,DLF_Score,width=width,color="skyblue",edgecolor="black",label="DLF_Score")
plt.bar(z,TATA_IPL,width=width,color="gray",edgecolor="black",label="TATA_IPL")
plt.xticks(x+width/2,names)    # if you names no center
#plt.xticks(x,names)      # if you give name for particuler bar that time this use


plt.title("Krishi_Bar",fontsize=20,color="red")
plt.xlabel("X-axis",fontsize=10,color="blue")
plt.ylabel("Y-axis",fontsize=10,color="green")
plt.legend(loc="upper left",framealpha=True,shadow=True,facecolor="yellow"
           ,edgecolor="black",title="BAR_GRAPH",title_fontsize=10,fancybox=True)
plt.show()
```
