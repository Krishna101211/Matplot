#**Marker**
marker=*,o,H,T,^,+  but mention in '' format

mhc=markers head colore(inside dot coloure)

mfc=marker face colore  

mew=marker edge width

mec=marker edge colore

```python
x=[1,2,3,4,5,6]
y=[10,15,20,25,30,35]
plt.plot(x,y,linestyle='solid',c='purple',
         marker='*',mfc='green',mew=15,mec='yellow')
plt.show()#if you want to see data then we go for show function
```

aplha handles transparency opacity its range is only between 0 to 1

```python
x=[1,2,3,4,5,6]
y=[10,15,20,25,30,35]
plt.plot(x,y,linestyle='solid',c='purple',
         marker='*',mfc='green',mew=15,mec='yellow',alpha=0.2)#alpha=transparency
plt.show()#if you want to see data then we go for show function
```

#**label**

```python
x=[1,2,3,4,5,6]
y=[10,15,20,25,30,35]
plt.plot(x,y,linestyle='solid',c='purple',
         marker='*',mfc='green',mew=15,mec='yellow',label='lineconnect')
plt.legend()#to show label
plt.title('Lineplot',fontsize=30)#to show title
plt.xlabel('x-axis',fontsize=20)#to show x-axis label
plt.ylabel('y-axis',fontsize=20)#to show y-axis label
plt.show()#if you want to see data then we go for show function
```

legend location -- to find values just give legend value error


-->ValueError:'upperright' is not a valid value for loc; supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'


```python
x=[1,2,3,4,5,6]
y=[10,15,20,25,30,35]
plt.plot(x,y,linestyle='solid',c='purple',
         marker='*',mfc='green',mew=15,mec='yellow',label='lineconnect')
plt.legend(loc="upperright")#to show label
plt.title('Lineplot',fontsize=30)#to show title
plt.xlabel('x-axis',fontsize=20)#to show x-axis label
plt.ylabel('y-axis',fontsize=20)#to show y-axis label
plt.show()#if you want to see data then we go for show function
```

```python
x=[1,2,3,4,5,6]
y=[10,15,20,25,30,35]
plt.plot(x,y,linestyle='solid',c='purple',
         marker='*',mfc='green',mew=15,mec='yellow',label='lineconnect')
plt.legend(loc="upper right",framealpha=False)#frame alpha used to remove box lineconnect
plt.title('Lineplot',fontsize=30)#to show title
plt.xlabel('x-axis',fontsize=20)#to show x-axis label
plt.ylabel('y-axis',fontsize=20)#to show y-axis label
plt.show()#if you want to see data then we go for show function
```

```python
x=[1,2,3,4,5,6]
y=[10,15,20,25,30,35]
plt.plot(x,y,linestyle='solid',c='purple',
         marker='*',mfc='green',mew=15,mec='yellow',label='lineconnect')
plt.legend(loc="upper right",framealpha=True)
plt.title('Lineplot',fontsize=30)#to show title
plt.xlabel('x-axis',fontsize=20)#to show x-axis label
plt.ylabel('y-axis',fontsize=20)#to show y-axis label
plt.show()#if you want to see data then we go for show function
```

```python
x=[1,2,3,4,5,6]
y=[10,15,20,25,30,35]
plt.plot(x,y,linestyle='solid',c='purple',
         marker='*',mfc='green',
         mew=15,mec='yellow',
         label='lineconnect')

plt.legend(loc="upper right",framealpha=True,
           shadow=True,
           facecolor='pink',
           title="Plotgraph",title_fontsize=15)#shadow use give shadow and facecolore=give the colore

plt.title('Lineplot',fontsize=30)#to show title
plt.xlabel('x-axis',fontsize=20)#to show x-axis label
plt.ylabel('y-axis',fontsize=20)#to show y-axis label
plt.show()#if you want to see data then we go for show function
```

#**grid()**

```python
x=[1,2,3,4,5,6]
y=[10,15,20,25,30,35]
plt.plot(x,y,linestyle='solid',c='purple',
         marker='*',mfc='green',
         mew=15,mec='yellow',
         label='lineconnect')

plt.legend(loc="upper right",framealpha=True,
           shadow=True,
           facecolor='pink',
           title="Plotgraph",title_fontsize=15,)#shadow use give shadow and facecolore=give the colore

plt.title('Lineplot',fontsize=30)#to show title
plt.xlabel('x-axis',fontsize=20)#to show x-axis label
plt.ylabel('y-axis',fontsize=20)#to show y-axis label

# plt.grid()#by default take horizontal and vertical
plt.grid(axis='y')#axis=x,both,y this parameters it take in ''format
plt.show()#if you want to see data then we go for show function
```

```python
x=[1,2,3,4,5,6]
y=[10,15,20,25,30,35]
plt.plot(x,y,linestyle='solid',c='purple',
         marker='o',mfc='green',
         mew=15,mec='yellow',
         label='lineconnect')

plt.legend(loc="upper right",framealpha=True,
           shadow=True,
           facecolor='pink',
           title="Plotgraph",title_fontsize=15,fancybox=True)#shadow use give shadow and facecolore=give the colore

plt.title('Lineplot',fontsize=30)#to show title
plt.xlabel('x-axis',fontsize=20)#to show x-axis label
plt.ylabel('y-axis',fontsize=20)#to show y-axis label

# plt.grid()#by default take horizontal and vertical
plt.grid(axis='both',c='green',ls='dotted',lw=2)#axis=x,both,y this parameters it take in ''format
plt.show()#if you want to see data then we go for show function
```
