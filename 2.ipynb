#scatter()

```python
x=[1,3,4,5,6,]
y=[2,4,6,8,10]
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=200)#size of line or dots
plt.show()
```

```python
plt.scatter(x,y,s=200,marker="*")
plt.show()
```

```python
plt.scatter(x,y,s=200,marker="*",c='Red')
plt.show()
```

```python
sizes=[1150,1200,1250,1300,1000]
plt.scatter(x,y,s=sizes,marker="*",c=['Red','Green','Blue','cyan','yellow'])#increase size of each element and changing color
plt.show()
```

```python
plt.title("Scatter Plot",fontsize=30,color="red")
plt.xlabel("X-axis",fontsize=20,color='blue')
plt.ylabel("Y-axis",fontsize=20)
sizes=[1150,1200,1250,1300,1000]
plt.scatter(x,y,s=sizes,marker="*",c=['Red','Green','Blue','cyan','yellow'])#increase size of each element and changing color
plt.show()
```

```python
plt.title("Scatter Plot",fontsize=30,color="red")
plt.xlabel("X-axis",fontsize=20,color='blue')
plt.ylabel("Y-axis",fontsize=20)
sizes=[150,120,125,130,100]

plt.scatter(x,y,s=sizes,marker="*",
            c=['Red','Green','Blue','cyan','yellow'],
            label='line',alpha=0.5
            )#to create legend we use label

plt.legend(loc='upper left',facecolor='yellow',edgecolor='black',shadow=True,fancybox=True)
plt.grid(True)
plt.show()
```

```python

sizes=[1150,1200,1250,1300,1000]

plt.title("Scatter Plot",fontsize=30,color="red")
plt.xlabel("X-axis",fontsize=20,color='blue')
plt.ylabel("Y-axis",fontsize=20)
sizes=[1150,1200,1250,1300,1000]

plt.scatter(x,y,s=sizes,marker="o",
            c=[1,2,3,4,5],cmap='plasma',
            label='Scatter_data_increase'
            )#to create legend we use label

plt.legend(loc='upper left',facecolor='yellow',edgecolor='black',shadow=True,fancybox=True)
plt.grid(True)
plt.show()
```

```python
a1=[1,2,3,4,5]
b1=[30,25,20,15,10]
sizes=[1150,1200,1250,1300,1000]
plt.scatter(x,y,s=sizes,marker="o",
            c=[1,2,3,4,5],cmap='inferno',
            label='Scatter_data_increase'
            )
plt.scatter(a1,b1,marker='o',s=100,c=[1,2,3,4,5],cmap='inferno_r') #colore_r will display colore in reversed sequence
plt.legend()
plt.show()
```

```python
a1=[1,2,3,4,5]
b1=[30,25,20,15,10]
sizes=[1150,1200,1250,1300,1000]
plt.scatter(x,y,s=sizes,marker="o",
            c=[1,2,3,4,5],cmap='inferno',
            label='Scatter_data_increase'
            )
plt.scatter(a1,b1,marker='o',s=100,c=[1,2,3,4,5],cmap='inferno_r')
plt.colorbar(label='Bar')
plt.legend()
plt.show()
```

#**pie chart**

```python
names=['java','python','c++','c']
values=[10,20,30,40]
plt.pie(values,labels=names,colors=['red','blue','yellow','green','c']
        ,explode=[0.0,0.7,0.0,0.0]#explode is used to take slice outside
        ,radius=1.5,#rdauis used to control diagram size
        labeldistance=1.2, # distance between diagram and label
        autopct="%1.2f%%",#use for display values in float format in diagram 0.2 will show 2 decimal values
        counterclock=True,
        startangle=180 ,#use for changing the angle of chart rotating chart
        textprops={"fontsize":10},
        wedgeprops={"linewidth":3,"width":1,"edgecolor":"black"}
        )
plt.legend(loc='lower left')
plt.show()
```
